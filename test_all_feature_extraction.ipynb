```python
import pandas as pd

test_asm_image = pd.read_csv("test_asm_image.csv")
test_4gram = pd.read_csv("test_data_750.csv")
test_dll = pd.read_csv("test_dll.csv")
test_hex = pd.read_csv("test_frequency.csv")
test_instr_count = pd.read_csv("test_instr_frequency.csv")
```

```python
merge1 = pd.merge(test_dll, test_asm_image, on="Id")
merge2 = pd.merge(merge1, test_hex, on="Id")
merge3 = pd.merge(merge2, test_4gram, on="Id")
test_all_features = pd.merge(merge3, test_instr_count, on="Id")
test_all_features
```

```python
test_all_features.to_csv("test_all_features.csv")
```
